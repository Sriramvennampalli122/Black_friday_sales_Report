```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly as pl
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("cleaned_BlackFriday_sales.csv")
df.drop("Unnamed: 0",inplace=True,axis=1)
```

```python
df
```

which occupation spend the most on purchases?

```python
df.groupby("Occupation").sum()["Purchase"].sort_values().plot(kind="bar",title="Total expenditure")
#sns.countplot(df["Occupation"])
```

from the Bar graph we can say that occupation 4 spend the most on purchases this can be attributed to the fact occupation 4 done the most number of purchases followed by 0,7 and so on.

Which occupation done the most number of purchases? 

```python
df.groupby("Occupation").size().sort_values().plot(kind="bar",title="Total expenditure")
```

from the Bar graph we can say that occupation 4 done the most number of purchases followed by 0,7 and so on.

which of Occupations has the highest mean expenditures on purchases?

```python
df.groupby("Occupation")["Purchase"].mean().sort_values().plot(kind="bar",title="Total expenditure")
```

From the graphs, we can conclude that occupation 17 the highest mean expendituer on purchases, closely followed by occupations 12 and 15. On the other hand, occupation 9 has the lowest mean expenditure.

List the Top 10  products that contributed the most to the income?

```python
df.groupby("Product_ID").sum()["Purchase"].nlargest(10).sort_values().plot(kind="bar")
```

The product that generated the highest income was P00025442, primarily because it was the most sold item. However, it is important to note that the lists of top-selling products and top income-generating products may not always align. This is due to the fact that a product with the highest sales volume could have a lower price point, which might result in less overall income compared to a higher-priced product with fewer sales
