```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly as pl
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("cleaned_BlackFriday_sales.csv")
df.drop("Unnamed: 0",inplace=True,axis=1)
```

```python
df.head()
```

which age group done the most number of purchases?

```python
df['Age'].value_counts()
```

```python
df.groupby('Age').size().plot(kind="bar",figsize=(12,6),title="Total purchases")
```

From the bar graph, it can be observed that the age category 26-35 accounts for the highest number of purchases. This can be attributed to the fact that individuals in this age group are typically employed, financially independent, and at a stage where they are beginning to get married. This is followed by the age groups 36-45, 18-25, 46-50, 51-55, 55+, and finally 0-17. The age group 0-17 makes the fewest purchases, likely because most individuals in this category are financially dependent on their parents.

who contributed more in the age group which done the highest number of purchases?

```python
sns.countplot(data=df,x="Age",hue="Gender")
```

From the bar graph, it can be observed that the age group 26-35 made the highest number of purchases. This can be attributed to the fact that most individuals in this group are employed and financially independent. Within this age group, males account for the majority of purchases, likely because employment rates are higher among men, while many women depend on male family members (husbands or fathers) for their needs. A similar trend is evident across all age groups.

Which age group in both the genders done the most number of purchases?

```python
sns.countplot(data=df,x="Gender",hue="Age")
```

The bar graph indicates that the age group 26-35 accounts for the majority of purchases across both genders. This can be attributed to the fact that most individuals in this group are employed and financially independent.

```python
df.groupby('Age')["Product_ID"].nunique()
```

Which age group purchased the widest variety of products sold?

```python
df.groupby('Age')["Product_ID"].nunique().plot(kind="bar",ylabel='No of products',title="Number of products purchased")
```

From the bar graph, it can be observed that the age group 26-35 purchased the widest variety of products. This can be attributed to various factors such as marriages and relocation to cities in search of employment. This is followed by the age groups 36-45, 18-25, 46-50, 51-55, 55+, and finally 0-17. The age group 0-17 purchased the least variety of products, as most buyers in this category are children who typically purchase general stationery items.

Which age  group contributed the most to the income of the company?

```python
df.groupby('Age').sum()["Purchase"].plot(kind="bar",ylabel="expenditure",title="Total expenditure")
```

By observing the bar graph, we can conclude that the age group 26-35 contributed the highest income, approximately 2 billion. This can be attributed to the fact that they made the largest number of purchases and bought a wide variety of products.

which age group spent most on purchases?

```python
df.groupby('Age')["Purchase"].mean().plot(kind="bar",ylabel="expenditure",title="Average expenditure")
```

```python
df.groupby('Age')["Purchase"].mean().plot(kind="pie",ylabel="expenditure",title="Average expenditure",autopct="%0.1f")
```
